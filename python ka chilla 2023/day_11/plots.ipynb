```python
#import data set
import seaborn as sns
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
# Count Plot
sns.countplot(x='survived', hue='sex', data=kashti)
```

```python
# Scatter Plot
sns.scatterplot(x='age', y='fare', hue='survived', data=kashti)
```

```python
sns.barplot(x='age', y='fare', hue='survived', data=kashti)
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10 , figsize=(10,10))
```

```python
kashti.columns
```

```python
sns.boxplot(x='sex', y='fare', hue='survived', data=kashti)
```

```python
sns.boxenplot(x='sex', y='fare', hue='survived', data=kashti)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(
    kashti,
    col='survived',
    row='pclass',
    hue='survived',
    margin_titles=True,
    palette={1: "seagreen", 0: "gray"}
)

g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
sns.stripplot(x='pclass', y='age', hue='survived', data=kashti, jitter=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns (very important)
corr = kashti.select_dtypes(include='number').corr()

plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', hue='survived', data=kashti, split=True)
```
