# EDA(Expolatory data analysis)
df: EDA is a method used to analyse and summarize dataset.In python you can use the pandas library to 
perform the eda on a dataset 

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset('titanic')
```

```python
# visualization in heatmap
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
```

```python
# Drop columns
df.drop(columns = 'deck', inplace=True)
```

```python
# visualize and explore 
sns.countplot(data = df, x = 'survived')
```

```python
sns.barplot(data =df, x ='survived', y = 'age', hue='who')
```

```python
sns.countplot(data = df, x = 'pclass')
```

```python
# Data distribution
sns.histplot(data = df, x = 'age', bins=8, kde=True)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.7)
```

```python
sns.boxplot(data = df, x = 'survived', y = 'age', hue='who')
```

```python
sns.scatterplot(data = df, x = 'age', y = 'fare',hue='pclass' )
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue='survived')
```

```python
# Stacked plot
df.groupby(['survived','pclass']).size().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data = df, x = 'sex', y = 'age', hue='who')
```

```python
# Point plot
sns.pointplot(data = df, x = 'pclass', y = 'fare', hue='who')
```

```python
# Relation of two numeric variables
sns.lmplot(
    data=df[df['fare'] < 300],
    y='fare',
    x='age',
    hue='sex'
)
```

```python
# How to add a column to make age groups
sns.histplot(data = df, x = 'age', bins=8, kde=True)
```

```python
# Add cloumn
df.columns
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0,12,20,40,60,80] , labels=['Child','Teenager','Adult','Middle Age','Senior'])
```
