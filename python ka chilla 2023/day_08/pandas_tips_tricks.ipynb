# 16- Spliting a String into Multiple Column

```python
df = pd.DataFrame({
    'name': ['Muhammad Ahmad', 'Ali Hassan', 'Ayesha Khan', 'Sajid Ali'],
    'Location': ['Lahore , Pakistan', 'Karachi , Pakistan', 'Islamabad , Pakistan', 'Multan , Pakistan']
})
df

```

```python
df.name.str.split(' ').head()
```

```python
# Spliting a column into two columns
df.name.str.split(' ',expand = True)
```

```python
df
```

```python
# Adding new columns after spliting
df[["first_name", "last_name"]] = df["name"].str.split(" ", expand=True).iloc[:, :2]
df
```

```python
df[['City', 'Country']] = df.Location.str.split(' , ', expand=True)   
df
```

```python
# Refine Data Manipulation

df[['first_name', 'last_name', 'City', 'Country']]
df
```

```python
fasla.head().T
```
